```python
root = 'https://api.opendota.com/api'
endpoint = '/proMatches'
```

```python
params = {'api_key': dota_api_key}
```

```python
requesting = requests.get(root + endpoint, params = params)
```

```python
result_proMatches = json.loads(requesting.text)
```

```python
df_proMatches = pd.json_normalize(result_proMatches)
```
